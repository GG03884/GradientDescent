```python
import numpy as np
from sklearn import datasets
```

```python
data=np.loadtxt("training_boston_x_y_train.csv",delimiter=",") ## loading training data
data.shape
X=data[:,0:13]
X1=np.append(X,np.ones(len(X)).reshape(-1,1),axis=1) ## Adding 1s for the intercept
Y=data[:,13]
```

## Function to update the values of m

```python
def stepgradient(X1,Y,learning_rate,m):
    m_slope=np.zeros(len(X1[0]))
    newslope=np.zeros(len(X1[0]))
    M=len(X1)    
    for i in range (M):
        y=Y[i]
        for j in range(len(X1[0])):
            m_slope[j]+=(-2/M)*(y-sum(m*X1[i]))*X1[i][j]      
    
        newslope=m-learning_rate*m_slope
    return newslope  
             
   
```

```python
def gd(X1,Y,learning_rate,num_iterations):
    m=np.zeros(len(X1[0]))
    for i in range(num_iterations):
        m=stepgradient(X1,Y,learning_rate,m)
        print(i," cost ",cost(X1,Y,m))
    return m    
```

## Cost function

```python
def cost(X1,Y,m):
    totalcost=0
    M=len(X1)
    for i in range(len(X1)):
        totalcost+=(1/M)*((Y[i]-sum(m*X1[i]))**2)
    return totalcost              
```

```python
def run():
    learning_rate=0.1
    num_iteration=300
    m=gd(X1,Y,learning_rate,num_iteration)
    print(m)
    return m
```

```python
m=run()
```

## Predict Function

```python
def predict(m,testing_data1):
    y_pred=[]
    for i in range(len(testing_data1)):
        ans=0;
        ans+=sum(m*testing_data1[i])
        y_pred.append(ans)
    return y_pred    
    
```

```python
testing_data=np.loadtxt("test_boston.csv",delimiter=",") ##loading testing data
testing_data1=np.append(testing_data,np.ones(len(testing_data)).reshape(-1,1),axis=1) ## Adding 1s column for the intercept
y_pred=predict(m,testing_data1)
np.savetxt("boston_y_pred.csv",y_pred,delimiter=",",fmt='%1.5f')
```
